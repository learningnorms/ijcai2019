# Convergence Plots

These are plots showing behavior of the agent's learning and convergence of the uncertainty interval over lifetimes.
These were simulations done separate from large scale simulations

Assumed that we have three norms A, B, C
A = -1 Forbidden, B = 0 Optional, C = 1 Obligatory



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Base case A_base
normA = pd.read_csv("single_run/A_base.csv")
normB = pd.read_csv("single_run/B_base.csv")
normC = pd.read_csv("single_run/C_base.csv")
```

```python
axA = normA.plot(color = 'r',linewidth=0.2,legend=False)  
axB = normB.plot( ax=axA,color = 'b',linewidth=0.2,legend=False)  
axC = normC.plot(ax=axA,color = 'g',linewidth=0.2,legend=False)

axA.fill_between(normA.index,normA.plausibility,normA.belief,facecolor='red',alpha=.1)
axB.fill_between(normB.index,normB.plausibility,normB.belief,facecolor='blue',alpha=.1)
axC.fill_between(normC.index,normC.plausibility,normC.belief,facecolor='green',alpha=.1)
axA.grid()

plt.ylim(0, 1)
plt.xlim(0,10)

```

```python
# Plotting them all together
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

# activate latex text rendering
rc('text', usetex=True)


# Base case A_base
normA = pd.read_csv("single_run/A_base.csv")
normB = pd.read_csv("single_run/B_base.csv")
normC = pd.read_csv("single_run/C_base.csv")

fig = plt.figure(figsize=(16,10))

# Basic case
basic = fig.add_subplot(221)
normA.plot(ax= basic, color = 'r',linewidth=0.2,legend=False)
normB.plot( ax=basic,color = 'b',linewidth=0.2,legend=False) 
normC.plot(ax=basic,color = 'g',linewidth=0.2,legend=False)
f = basic.fill_between(normA.index,normA.plausibility,normA.belief,facecolor='red',alpha=.1)
p = basic.fill_between(normB.index,normB.plausibility,normB.belief,facecolor='blue',alpha=.1)
o = basic.fill_between(normC.index,normC.plausibility,normC.belief,facecolor='green',alpha=.1)
basic.grid()
basic.set_ylim(0, 1)
basic.set_xlim(0,10)
#basic.set_xlabel('Observations', fontsize=20)
#basic.set_ylabel('Uncertainty', fontsize=20)
basic.legend(handles = [f,p,o], labels=["Forbidden","Optional","Obligatory"],loc='upper right', fontsize='x-large')
basic.set_title('Compliance: High, Sanctioning: High, Ambiguity: Low', fontsize=20)


# Sanction Case
normA_nosanc = pd.read_csv("single_run/A_nosanc.csv")
normB_nosanc = pd.read_csv("single_run/B_nosanc.csv")
normC_nosanc = pd.read_csv("single_run/C_nosanc.csv")

sanction = fig.add_subplot(222, sharex = basic, sharey = basic)
normA_nosanc.plot(ax= sanction, color = 'r',linewidth=0.2,legend=False)
normC_nosanc.plot(ax=sanction,color = 'g',linewidth=0.2,legend=False)
f = sanction.fill_between(normA_nosanc.index,normA_nosanc.plausibility,normA_nosanc.belief,facecolor='red',alpha=.1)
o = sanction.fill_between(normC_nosanc.index,normC_nosanc.plausibility,normC_nosanc.belief,facecolor='green',alpha=.1)
sanction.grid()
sanction.set_ylim(0, 1)
sanction.set_xlim(0,10)
#sanction.set_xlabel('Observations', fontsize=20)
#sanction.set_ylabel('Uncertainty', fontsize=20)
sanction.legend(handles = [f,o], labels=["Forbidden","Obligatory"],loc='upper right', fontsize='x-large')
sanction.set_title('Compliance: High, Sanctioning: '+ r"\textbf{Low}"+ ', Ambiguity: Low', fontsize=20)


# NonCompliance Case
normA_nocomp = pd.read_csv("single_run/A_nocomp.csv")
normB_nocomp = pd.read_csv("single_run/B_nocomp.csv")
normC_nocomp = pd.read_csv("single_run/C_nocomp.csv")

compliance = fig.add_subplot(223, sharex = basic, sharey = basic)
normA_nocomp.plot(ax= compliance, color = 'r',linewidth=0.2,legend=False)
normC_nocomp.plot(ax=compliance,color = 'g',linewidth=0.2,legend=False)
f = compliance.fill_between(normA_nocomp.index,normA_nocomp.plausibility,normA_nocomp.belief,facecolor='red',alpha=.1)
o = compliance.fill_between(normC_nocomp.index,normC_nocomp.plausibility,normC_nocomp.belief,facecolor='green',alpha=.1)
compliance.grid()
compliance.set_ylim(0, 1)
compliance.set_xlim(0,10)
#compliance.set_xlabel('Observations', fontsize=20)
#compliance.set_ylabel('Uncertainty', fontsize=20)
compliance.legend(handles = [f,o], labels=["Forbidden","Obligatory"],loc='upper right', fontsize='x-large')
compliance.set_title('Compliance: '+ r"\textbf{Low}"+ ', Sanctioning: '+ r"\textbf{Low}"+ ', Ambiguity: Low', fontsize=20)

# Ambiguity
normA_am = pd.read_csv("single_run/A_am.csv")
normB_am = pd.read_csv("single_run/B_am.csv")
normC_am = pd.read_csv("single_run/C_am.csv")

ambiguity = fig.add_subplot(224,sharex = basic, sharey = basic)
normA_am.plot(ax=ambiguity, color = 'r',linewidth=0.2,legend=False)
normC_am.plot(ax=ambiguity,color = 'g',linewidth=0.2,legend=False)
f = ambiguity.fill_between(normA_am.index,normA_am.plausibility,normA_am.belief,facecolor='red',alpha=.1)
o = ambiguity.fill_between(normC_am.index,normC_am.plausibility,normC_am.belief,facecolor='green',alpha=.1)
ambiguity.grid()
ambiguity.set_ylim(0, 1)
ambiguity.set_xlim(0,10)
#ambiguity.set_xlabel('Observations', fontsize=20)
#ambiguity.set_ylabel('Uncertainty', fontsize=20)
ambiguity.legend(handles = [f,o], labels=["Forbidden","Obligatory"],loc='upper right', fontsize='x-large')
ambiguity.set_title('Compliance: High, Sanctioning: High, Ambiguity: '+ r"\textbf{High}", fontsize=20)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.15,
                    wspace=0.1)

fig.text(0.54, 0.02, 'Number of Observations', ha='center', fontsize=25)
fig.text(0.04, 0.5, 'Uncertainty Interval Bounds: [Belief, Plausibility]', va='center', rotation='vertical', fontsize=25)

fig.savefig('test.png')
```
